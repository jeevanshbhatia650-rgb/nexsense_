```python

# 1️⃣ Source file path (from your drive)
source_path = r"C:\Users\JEEVANSH\Downloads\Dataset.csv"

# 2️⃣ Read the file
df = pd.read_csv(source_path)

# 3️⃣ Define subfolder name
subfolder = "fruits_data"

# 4️⃣ Create subfolder if it doesn't exist
os.makedirs(subfolder, exist_ok=True)

# 5️⃣ Save inside subfolder
destination_path = os.path.join(subfolder, "dataset.csv")

df.to_csv(destination_path, index=False)

print("File saved inside subfolder successfully!")
```

```python
df.iloc[:, 2:5] = df.iloc[:, 2:5].apply(pd.to_numeric, errors='coerce')
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
string_cols = df.select_dtypes(include=['object']).columns
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill string with mode
df[string_cols] = df[string_cols].fillna(df[string_cols].mode().iloc[0])
```

```python
import random

def generate_expiry(row):
    base = 10
    
    # reduce expiry if temperature is high
    temp_factor = (row["Temp"] - 20) * 0.5
    
    # reduce expiry if humidity is high
    humid_factor = (row["Humid (%)"] - 80) * 0.2
    
    # random noise
    noise = random.uniform(-1, 1)
    
    expiry = base - temp_factor - humid_factor + noise
    
    return max(1,round(expiry))
df["Expiry_Days"] = df.apply(generate_expiry, axis=1)

df.head()
```
